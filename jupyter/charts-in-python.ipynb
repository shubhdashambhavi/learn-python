```python
counts = Calls["job"].value_counts()
counts
```

```python
import seaborn as sns

plt_1= counts.plot(kind='bar',
                  title =" Job Types",
                  figsize =(10,5),
                  color= sns.color_palette("BuGn_r", len(counts)),
                  fontsize=10)
plt_1.set_xlabel("Job description")
plt_1.set_ylabel("Frequence")
plt_1
```

```python
# Horizontal bar
import seaborn as sns

plt_1= counts.plot(kind='barh',
                  title =" Job Types",
                  figsize =(10,5),
                  color= sns.color_palette("BuGn_r", len(counts)),
                  fontsize=10)
plt_1.set_xlabel("Job description")
plt_1.set_ylabel("Frequence")
plt_1
```
